# Linear Regression with Sci-Kit Learn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

## Linear Regression Function

```python
def LR(x,w,b):
    return np.dot(x,w) + b
```

## Loading the data from the dataset

```python
dataset = datasets.load_diabetes()
scaler = StandardScaler()


x_train = dataset.data
nx_train = scaler.fit_transform(x_train) #Normalizes the features data
x_features = ["Age","Sex","Body Mass Index","Blood Pressure","Serum Cholesterol",
              "Low Lipoproteins","High Lipoproteins","Total Cholesterol", "Triglycerides","Blood Sugar Level"]
y_train = dataset.target
epochs = 10000

m,n = nx_train.shape

#print(x_train)
#print(nx_train)
#print(y_train)
```

## Model Fit

```python
model = SGDRegressor(max_iter=epochs)
print(model)
model.fit(nx_train,y_train)
print("Iterations Completed: ",model.n_iter_)
print("Weights updated:", model.t_)
```

```python
w = model.coef_
b = model.intercept_

print(f"Weights:{w}\n Biases:{b}")
```

## Predictions with both(Implemented LR and Sci-Kit Model)

```python
sgd_predicts = model.predict(nx_train)
predicts = LR(nx_train,w,b)
#print(y_train)
#print(sgd_predicts)
#print(predicts)
```

## Ploting the predictions

```python
fig,ax = plt.subplots(1,n, figsize=(30,3),sharey=True)
for i in range(len(ax)):
    ax[i].scatter(nx_train[:,i],y_train, )
    ax[i].scatter(nx_train[:,i],sgd_predicts)
    ax[i].set_xlabel(x_features[i])

fig.suptitle("target versus prediction using z-score normalized model")
plt.show()
```
