# Linear Regression Model 
### With MSE cost function and Gradiend Descent)

## -----------------------------------------------------------

## First, we set our dataset and variables

```python
import numpy as np
import matplotlib.pyplot as plt


x_train = np.array([1,3,1,2,4,5,3,7,9,8,9,4,3,8,5,6,7,8]) 
y_train = np.array([200,500,200,400,800,700,300,900,100,200,900,400,500,800,500,600,700,800])

w = np.random.rand() * 500   # Weight
b = np.random.rand() * 300  # Bias

print("w: ",w)
print("b: ",b)

m = len(x_train)
num_iter = 10000
alpha = 0.04

```

## Graph for the x/y values in our data set

```python
plt.scatter(x_train,y_train)

plt.show()
```

## Simple Linear Regression

```python
def simpleLR(x,w,b,m):
    predictions = []
    
    for i in range(m):
        fw_b = w * x[i] + b
        predictions.append(fw_b)
    
    return predictions
```

## Cost Function 

```python
def computeCost(x,y,w,b,m):
    costPred = []
    costSum = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        costSum += cost
        costPred.append(cost)
        
    totalCost = (1/(2 * m)) * costSum
    
    return totalCost
```

## Gradient Function

```python
def computeGradient(x,y,w,b,m):
    aux_w = 0
    aux_b = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        aux_w_i = (f_wb - y[i]) * x[i]
        aux_b_i = f_wb - y[i]
        
        aux_w += aux_w_i
        aux_b += aux_b_i
        
    aux_w = (1 / m) * aux_w
    aux_b = (1 / m) * aux_b
    
    return aux_w, aux_b
    
```

## Gradient Descent 


```python
def GradientDescent(alpha,x,y,initial_w,initial_b,m,num_iter):
   
    cost_history = []
    
    w = initial_w
    b = initial_b
    
    
    for i in range(num_iter):
        aux_w, aux_b = computeGradient(x,y,w,b,m)
        
        b = b - alpha * aux_b
        w = w - alpha * aux_w
        
        cost_history.append(computeCost(x,y,w,b,m))
        
    return cost_history,w,b
    
```

## Without Gradient Descent Optimization

```python
predictions = simpleLR(x_train,w,b,m)
cost = computeCost(x_train,y_train,w,b,m)

plt.scatter(x_train,y_train)
plt.plot(x_train,predictions)
plt.xlim(0,10)
plt.ylim(0,1000)
plt.text(0,20,f'Cost: {cost:.2f}')

plt.show()
```

## After Gradient Descent Optimization

```python
cost_history,final_w,final_b = GradientDescent(alpha,x_train,y_train,w,b,m,num_iter)

predictions = simpleLR(x_train,final_w,final_b,m)
cost = computeCost(x_train,y_train,final_w,final_b,m)



plt.scatter(x_train,y_train)
plt.plot(x_train,predictions)
plt.xlim(0,10)
plt.ylim(0,1000)
plt.text(0,20,f'Cost: {cost:.2f}')

plt.show()
```
